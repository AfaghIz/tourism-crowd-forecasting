```python
import pandas as pd

data = pd.read_csv("../data/processed/weekly_crowd_index.csv", parse_dates=["date"])
```

```python
#get time features
data["month"] = data["date"].dt.month
data["week_of_year"] = data["date"].dt.isocalendar().week.astype(int)
```

```python
#addin seasons
def get_season(m):
    if m in [12, 1, 2]:
        return "winter"
    elif m in [3, 4, 5]:
        return "spring"
    elif m in [6, 7, 8]:
        return "summer"
    else:
        return "autumn"

data["season"] = data["month"].apply(get_season)
```

```python
#encoding season bc ML needs numbers
data = pd.get_dummies(data, columns=["season"], drop_first=True)
```

```python
data.to_csv("../data/processed/model_dataset.csv", index=False)
data.head() #saving
```
