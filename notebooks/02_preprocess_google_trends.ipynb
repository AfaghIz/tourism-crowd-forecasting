```python
import pandas as pd

BASE_PATH = "/Users/Afagh/Downloads/"

files = {
    "travel_intent": "STVisit_Istanbul.csv",
    "attractions": "STIstanbul_tourist_attractions.csv",
    "museum": "STIstanbul_museums.csv",
    "historic": "STIstanbul_historical_sites.csv",
    "park": "STIstanbul_parks.csv"
}

dfs = []

for category, file in files.items():
    # Read CSV, skip metadata rows automatically
    df = pd.read_csv(BASE_PATH + file, skiprows=1)

    # First column = date (Week or Month)
    date_col = df.columns[0]

    # Second column = interest value
    value_col = df.columns[1]

    df = df[[date_col, value_col]]
    df.columns = ["date", "trend_score"]

    df["category"] = category
    dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)

# Convert to datetime
final_df["date"] = pd.to_datetime(final_df["date"])

final_df.to_csv("../data/raw/google_trends_istanbul.csv", index=False)

print(final_df.head())
print(final_df.category.value_counts())
```
