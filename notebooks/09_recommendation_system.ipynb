```python
import pandas as pd

data = pd.read_csv("../data/processed/predictions.csv")
pois = pd.read_csv("../data/processed/pois_istanbul_clean.csv")

print(data.columns)
print(pois.columns)
```

```python
today = data.iloc[0]
current_crowd = today["predicted_crowd"]

print("Current crowd level:", current_crowd)
```

```python
def distance(lat1,lon1,lat2,lon2):
    return ((lat1-lat2)**2 + (lon1-lon2)**2) ** 0.5

#user location (example Sultanahmet)
user_lat = 41.0082
user_lon = 28.9784

pois["distance"] = pois.apply(
    lambda row: distance(user_lat, user_lon, row["lat"], row["lon"]), axis=1
)
```

```python
pois["score"] = current_crowd * 0.6 + pois["distance"] * 0.4
```

```python
def crowd_label(x):
    if x < 0.3:
        return "Low"
    elif x < 0.7:
        return "Medium"
    return "High"

print("\nCurrent crowd level:", crowd_label(current_crowd))
```

```python
recommendations = pois.sort_values("score").head(10)

print("\nTop Recommended Places:\n")

for i, row in recommendations.iterrows():
    print(f"- {row['name']} ({row['category']}) | Distance: {row['distance']:.4f}")
```
