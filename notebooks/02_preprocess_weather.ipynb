```python
import pandas as pd

weather = pd.read_csv("../data/raw/weather_istanbul_daily.csv", parse_dates=["date"])
weather.head()
```

```python
weather_weekly = weather.set_index("date").resample("W").agg({
    "temp_max": "mean",
    "temp_min": "mean",
    "temp_avg": "mean",
    "precipitation": "sum"
}).reset_index()

weather_weekly.head()
```

```python
weather_weekly.to_csv("../data/processed/weather_istanbul_weekly.csv", index=False)
print(weather_weekly.shape)
```
