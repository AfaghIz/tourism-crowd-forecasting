```python
import pandas as pd

pois = pd.read_csv("../data/raw/pois_istanbul.csv")
pois.head()
```

```python
#getting rid of POIs with no name and removing dupes
pois = pois[pois["name"] != "Unknown"]
pois = pois.drop_duplicates(subset=["name", "lat", "lon"])
```

```python
#simplifying catergories
def simplify_category(cat):
    cat = str(cat).lower()

    if "museum" in cat:
        return "museum"
    elif "historic" in cat:
        return "historic"
    elif "park" in cat:
        return "park"
    elif "attraction" in cat:
        return "attraction"
    else:
        return "other"

pois["category_clean"] = pois["category"].apply(simplify_category)
```

```python
#keeping touirism relevant poi's 
pois = pois[pois["category_clean"] != "other"]
#saves
pois.to_csv("../data/processed/pois_istanbul_clean.csv", index=False)
print("Final POIs:", pois.shape)
pois["category_clean"].value_counts()
```
